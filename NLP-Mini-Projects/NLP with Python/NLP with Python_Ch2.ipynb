```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

```python
# Compare genres in their usage of model verbs
from nltk.corpus import brown
news_text = brown.words(categories='news')
fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m], end = ' ')
```

```python
# 1.5 Inaugural Address Corpus
from nltk.corpus import inaugural
inaugural.fileids()
```

```python
[fileid[:4] for fileid in inaugural.fileids()]
```

```python
#Converts the words in the Inauaugal corpus to lowercaes using w.lower(), then check if start with america or citizen
cfd = nltk.ConditionalFreqDist(
        (target, fileid[:4])
        for fileid in inaugural.fileids()
        for w in inaugural.words(fileid)
        for target in ['america', 'citizen']
        if w.lower().startswith(target))
```

```python
cfd.plot()
```

```python
# 2.3 Plotting and Tabulating Distibutions
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist(
        (target, fileid[:4])
        for fileid in inaugural.fileids()
        for w in inaugural.words(fileid)
        for target in ['america', 'citizen']
        if w.lower().startswith(target))
```

```python
#Condition is the name of language and the counts from word lengths
from nltk.corpus import udhr
languages = ['Chickasaw', 'English', 'German_Deutsch', 'Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
        (lang, len(word))
        for lang in languages
        for word in udhr.words(lang + '-Latin1'))
```

```python
cfd.tabulate(conditions=['English', 'German_Deutsch'],
            samples=range(10), cumulative=True)
```

```python
def generate_model(cfdist, word, num=15):
    for i in range(num):
        print(word, end = ' ')
        word = cfdist[word].max()

text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
```

```python
cfd['living']
```

```python
generate_model(cfd, 'living')
```
